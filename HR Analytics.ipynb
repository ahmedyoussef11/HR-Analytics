```python
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline 

import warnings
warnings.filterwarnings('ignore')

import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\HP\\Downloads\\Data P3 MeriSKILL\\HR-Employee-Attrition.csv")
df
```

```python
df.head(10)
```

```python
df.drop(['StandardHours', 'EmployeeCount'], axis=1, inplace=True)
df
```

```python
df.isnull().sum()
```

```python
correlation_matrix = df.corr()
correlation_matrix
```

```python
plt.figure(figsize = (14,14))
sns.heatmap(correlation_matrix, annot=True, cmap='Purples', fmt=".2f")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.hist(df['OverTime'], bins=20, edgecolor='k')
plt.title('Distribution of Overtime')
plt.xlabel('Overtime Hours')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(x='MaritalStatus', data=df, palette='Set1')
plt.title('Marital Status Distribution')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='JobRole', data=df, palette='Set2')
plt.title('Job Role Distribution')
plt.xlabel('Job Role')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(5, 5))
df['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='EducationField', data=df, palette='viridis')
plt.title('Education Field Distribution')
plt.xlabel('Education Field')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(x='Department', data=df, palette='pastel')
plt.title('Department Distribution')
plt.xlabel('Department')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(x='BusinessTravel', data=df, palette='dark')
plt.title('Business Travel Distribution')
plt.xlabel('Business Travel')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.scatter(df['OverTime'], df['Age'], alpha=0.5)
plt.title('Relationship between Overtime and Age')
plt.xlabel('Age')
plt.ylabel('Overtime Hours')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.hist(df['TotalWorkingYears'], bins=20, edgecolor='k')
plt.title('Distribution of Total Years Worked')
plt.xlabel('Total Years Worked')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(x='Education', data=df, palette='muted')
plt.title('Educational Level Distribution')
plt.xlabel('Educational Level')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.hist(df['NumCompaniesWorked'], bins=20, edgecolor='k')
plt.title('Distribution of Number of Operating Companies')
plt.xlabel('Number of Operating Companies')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['DistanceFromHome'], bins=20, edgecolor='k')
plt.title('Distribution of Distance from Home')
plt.xlabel('Distance from Home')
plt.ylabel('Frequency')
plt.show()
```
